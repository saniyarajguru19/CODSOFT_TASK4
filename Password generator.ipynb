```python
import tkinter as tk
from tkinter import messagebox
import random
import string

# Function to generate the password
def generate_password():
    try:
        length = int(entry_length.get())
        if length <= 0:
            messagebox.showerror("Error", "Please enter a positive number for the length.")
            return
        
        # Define the possible characters
        characters = string.ascii_letters + string.digits + string.punctuation
        
        # Generate the password
        password = ''.join(random.choice(characters) for _ in range(length))
        
        # Display the password
        result_label.config(text=f"Generated Password: {password}")
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid number for the length.")

# Function to reset the input and output fields
def reset():
    entry_length.delete(0, tk.END)
    result_label.config(text="Generated Password:")

# Create the main window
root = tk.Tk()
root.title("Password Generator")

# Set background color
root.configure(bg="#2c3e50")

# Create labels
title_label = tk.Label(root, text="Password Generator", font=("Arial", 20), bg="#2c3e50", fg="#ecf0f1")
title_label.grid(row=0, column=0, columnspan=2, pady=10)

length_label = tk.Label(root, text="Enter password length:", font=("Arial", 14), bg="#2c3e50", fg="#ecf0f1")
length_label.grid(row=1, column=0, padx=10, pady=5, sticky="e")

# Create entry field
entry_length = tk.Entry(root, font=("Arial", 14), bg="#ecf0f1", fg="#2c3e50")
entry_length.grid(row=1, column=1, padx=10, pady=5)

# Create buttons
generate_button = tk.Button(root, text="Generate", command=generate_password, bg="#3498db", fg="#000000", font=("Arial", 12))
generate_button.grid(row=2, column=0, columnspan=2, pady=10)

reset_button = tk.Button(root, text="Reset", command=reset, bg="#e74c3c", fg="#000000", font=("Arial", 12))
reset_button.grid(row=3, column=0, columnspan=2, pady=10)

# Create result label
result_label = tk.Label(root, text="Generated Password:", font=("Arial", 14), bg="#2c3e50", fg="#ecf0f1")
result_label.grid(row=4, column=0, columnspan=2, pady=10)

# Run the main event loop
root.mainloop()
```
